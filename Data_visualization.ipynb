### **Import Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px


pd.set_option('display.max_columns', None)  
pd.set_option('display.width', None)
```

```python
df = pd.read_csv("Spotify Most Streamed Songs.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
```

```python
df = df.dropna(subset=['streams'])
```

```python
df.head(5)
```

```python
df_filtered = df[(df['released_year'] >= 2015) & (df['released_year'] <= 2020)]

artist_year_streams = df_filtered.groupby(['artist(s)_name', 'released_year'])['streams'].sum().unstack(fill_value=0)

top_artists = artist_year_streams.sum(axis=1).nlargest(5).index
top_artist_streams = artist_year_streams.loc[top_artists]

top_artist_streams.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Top 5 Artists by Total Streams (2015-2020)', fontsize=24)
plt.xlabel('Artist Name', fontsize=20)
plt.ylabel('Total Streams', fontsize=20)
plt.xticks(rotation=45, fontsize=18)
plt.yticks(fontsize=18)
plt.legend(title='Release Year', fontsize=16, title_fontsize=18)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.scatterplot(data=df,x='released_year', y='streams', hue='key', s = 100)
plt.title('Spotify Streams vs Year Released', fontsize=15, fontweight='bold')
plt.xlabel('Released Year', fontsize=12, fontweight = 'bold')
plt.ylabel(f'Streams', fontsize=12, fontweight = 'bold')
plt.grid(True, linestyle='--', alpha=0.9)

print('Scatter plot for released per year vs. Spotify Streams')
plt.show()
```

```python
df['released_year'] = pd.to_numeric(df['released_year'], errors='coerce').fillna(0).astype(int)
df['streams'] = pd.to_numeric(df['streams'], errors='coerce').fillna(0)

temporal_data = df.groupby('released_year')['streams'].sum()

plt.figure(figsize=(10, 6))
plt.plot(temporal_data.index, temporal_data.values, marker='o', linestyle='-', color='b', alpha=0.8)
plt.title('Figure 4: Song Popularity Trends Over Time', fontsize=16)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Total Streams (Billions)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

df['decade'] = (df['released_year'] // 10) * 10


streams_by_decade = df.groupby('decade')['streams'].sum()


plt.figure(figsize=(12, 6))
streams_by_decade.plot(kind='bar', color='steelblue', edgecolor='black', alpha=0.8)
plt.title('Figure 3: Total Streams by Decade (1930-2023)', fontsize=24)
plt.xlabel('Decade', fontsize=20)
plt.ylabel('Total Streams', fontsize=20)
plt.xticks(rotation=45, fontsize=18)
plt.yticks(fontsize=18)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

df['bpm_range'] = pd.cut(df['bpm'], bins=range(0, 201, 20))
streams_by_bpm = df.groupby('bpm_range')['streams'].sum()


plt.figure(figsize=(12, 6))
streams_by_bpm.plot(kind='bar', color='salmon', edgecolor='black', alpha=0.8)
plt.title('Figure 5: Total Streams by BPM Range', fontsize=24)
plt.xlabel('BPM Range', fontsize=20)
plt.ylabel('Total Streams', fontsize=20)
plt.xticks(rotation=45, fontsize=18)
plt.yticks(fontsize=18)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

df['decade'] = (df['released_year'] // 10) * 10
bpm_danceability = df.groupby(['bpm_range', 'decade'])['danceability_%'].mean().reset_index()
bpm_danceability_cleaned = bpm_danceability.dropna(subset=['danceability_%'])

bpm_danceability_cleaned['bpm_range'] = bpm_danceability_cleaned['bpm_range'].astype(str)

fig = px.scatter(
    bpm_danceability_cleaned,
    x='decade',
    y='danceability_%',
    color='bpm_range',
    size='danceability_%',  
    hover_data={
        "bpm_range": True,  
        "danceability_%": ":.2f",  
        "decade": True      
    },
    labels={
        "danceability_%": "Danceability (%)",
        "decade": "Decade",
        "bpm_range": "BPM Range"
    },
    title="Figure 6: Danceability by BPM Range and Decade",
    template="plotly",
)


fig.update_traces(
    hovertemplate="<b>Decade:</b> %{x}<br>" +
                  "<b>BPM Range:</b> %{customdata[0]}<br>" +
                  "<b>Danceability:</b> %{y:.2f}%<br>"
)


fig.show()
```

```python

top_15_artists = df['artist(s)_name'].value_counts().head(15)
df['key'].replace('', 'Unknown', inplace=True)  
df['key'].fillna('Unknown', inplace=True)  
key_counts = df['key'].value_counts()


plt.figure(figsize=(16, 16))  


outer_colors = plt.cm.tab20(range(len(top_15_artists)))
outer_wedges, outer_texts, outer_autotexts = plt.pie(top_15_artists, radius=1.3, autopct='%1.1f%%',
                                                    startangle=90, colors=outer_colors,
                                                    pctdistance=0.85, wedgeprops=dict(width=0.3, edgecolor='w'))


bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"), bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(outer_wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    plt.annotate(top_15_artists.index[i], xy=(x, y), xytext=(1.4*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)


inner_colors = plt.cm.Set3(np.arange(len(key_counts)))
inner_wedges, _ = plt.pie(key_counts, radius=1.0, startangle=90,
                          colors=inner_colors, wedgeprops=dict(width=0.3, edgecolor='w'))


for i, wedge in enumerate(inner_wedges):
    angle = (wedge.theta1 + wedge.theta2) / 2
    center = wedge.center
    r = wedge.r
    x = center[0] + r * 0.7 * np.cos(np.deg2rad(angle))
    y = center[1] + r * 0.7 * np.sin(np.deg2rad(angle))
    plt.text(x, y, f"{key_counts.index[i]}\n({key_counts.values[i]})", ha='center', va='center', fontsize=9, rotation=angle-90)



plt.title('Figure 7: Top 15 Artists and Key Distribution', fontsize=16)


plt.axis('equal')
plt.show()
```
